# Importing Libraries

```python
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
 import seaborn as sns
```

Get The Data

```python
df=pd.read_csv('/content/drive/MyDrive/Data Analytic Project/Mall_Customers.csv')
```

Check the head of the data

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df.drop(columns=['CustomerID','Age','Gender'],axis=1).values
```

```python
X
```

```python
X.shape
```

Create a scatter plot

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=X[:,0],y=X[:,1])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=10)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=X[:,0],y=X[:,1], c=kmeans.labels_,cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('spending Score (1-100)')
plt.title('kmean')
plt.show()
```
